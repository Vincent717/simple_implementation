```python
import random
import numpy as np
import math

class Perceptron(object):
    
    sgn = lambda self, x: 1 if x >= 0 else -1
    sigmoid = lambda self, x: 1 / (1 + math.e**(-x))
    f = sgn
    
    
    alpha = 0.3
    bound = 500
    degree = 0
    inputs = np.array([])
    outputs = np.array([])
    weights = np.array([])
    bias = 0
    
    def initial(self, inputs, outputs):
        
        if len(inputs) != len(outputs):
            print('input set and output set have different size')
        else:
            self.inputs = np.array(inputs)
            self.size = len(inputs)
            self.degree = len(inputs[0])
            self.outputs = np.array(outputs)
        
        
    def train(self, type_of_learning_algorithm, alpha):
        # randomly initailize weights:
        weights = np.array([random.random() for i in range(0,self.degree)])
        bias    = random.random()
        
        delta_w = np.zeros(self.degree) 
        delta_b = 0
        
        #ã€€perceptron : with sign function
        if type_of_learning_algorithm == 'perceptron':
            f = self.sgn
            self.f = f
            print('training...')
            for k in range(0, self.bound):
                print('Epso %d' % k)
                i = random.randrange(0,self.size,1)
                print(weights)
                net = f(np.dot(self.inputs[i], weights) + bias)
                delta_w = self.alpha * (self.outputs[i] - net) * self.inputs[i]
                delta_b = self.alpha * (self.outputs[i] - net)
                weights = weights + delta_w
                bias    = bias    + delta_b   
                
        # delta : with sigmoid function as default
        elif type_of_learning_algorithm == 'delta':
            f = self.sigmoid
            self.f = f
            print('training...')
            for k in range(0, self.bound):
                print('Epso %d' % k)
                i = random.randrange(0,self.size,1)
                print(weights)
                net = np.dot(self.inputs[i], weights) + bias
                delta_w = self.alpha * (self.outputs[i] - f(net)) * self.inputs[i] * f(net)*(1-f(net))
                delta_b = self.alpha * (self.outputs[i] - f(net)) * f(net)*(1-f(net))
                weights = weights + delta_w
                bias    = bias    + delta_b

        self.weights = weights
        self.bias    = bias
    
    
    def feed(self, x):
        if len(x) != self.degree:
            print('Please enter an input with %d degree.' % self.degree)
        else:
            return self.f(np.dot(np.array(x), self.weights) + self.bias)
        
```

```python
t = Perceptron()
x = [[0,0],[0,1],[1,0],[1,1]]
y = [0,0,0,1]
t.initial(x,y)
t.train('perceptron',0.3)
np.array([random.random() for i in range(0,3)])
```
